```python
# Summary in one block:
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

file_path = '/content/drive/My Drive/DA_Practise_Folder_EXCEL/global_student_migration.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
# Check the structure, data types, and missing values
# Show info about data types, non-null counts, memory usage
df.info()

# Show count of missing values per column
print("\nMissing values per column:")
print(df.isnull().sum())

# Show first 5 rows again to remind the data structure
df.head()
```

```python
# fill those missing values in placement_country and placement_company with "Not Placed
# fill language_proficiency_test missing values with "None"
# Fill missing placement details with 'Not Placed'
df['placement_country'].fillna('Not Placed', inplace=True)
df['placement_company'].fillna('Not Placed', inplace=True)

# Fill missing language proficiency test with 'None'
df['language_proficiency_test'].fillna('None', inplace=True)

# Double-check if missing values remain
print(df.isnull().sum())
```

```python
# Summary Statistics for Numeric Columns
# Counts for Categorical columns
# Visualize placement status distribution - Target Variable

import matplotlib.pyplot as plt
import seaborn as sns

# Numeric summary stats
print("Numeric columns summary:\n", df.describe())

# Count of unique values in categorical columns (just a few examples)
print("\nValue counts for placement_status:\n", df['placement_status'].value_counts())
print("\nValue counts for scholarship_received:\n", df['scholarship_received'].value_counts())
print("\nValue counts for language_proficiency_test:\n", df['language_proficiency_test'].value_counts())

# Plot placement_status distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='placement_status')
plt.title('Placement Status Distribution')
plt.show()
```

```python
# Compare numeric features by placement status — e.g., do placed students have higher GPA or salary
import matplotlib.pyplot as plt
import seaborn as sns

numeric_features = ['gpa_or_score', 'starting_salary_usd', 'test_score']

plt.figure(figsize=(16, 5))

for i, feature in enumerate(numeric_features, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(x='placement_status', y=feature, data=df)
    plt.title(f'{feature} by Placement Status')
    plt.tight_layout()

plt.show()
```

```python
#  Calculate placement rate per category
# Function to calculate placement rate for a given column
def placement_rate(df, col):
    placement_counts = df.groupby([col, 'placement_status']).size().unstack(fill_value=0)
    placement_counts['placement_rate'] = placement_counts['Placed'] / (placement_counts['Placed'] + placement_counts['Not Placed'])
    return placement_counts.sort_values('placement_rate', ascending=False)

# Placement rates by origin_country
origin_placement = placement_rate(df, 'origin_country').reset_index()

# Placement rates by destination_country
destination_placement = placement_rate(df, 'destination_country').reset_index()

# Placement rates by university_name (top 15 by total students)
uni_counts = df['university_name'].value_counts().head(15).index
uni_placement = placement_rate(df[df['university_name'].isin(uni_counts)], 'university_name').reset_index()
```

```python
# Plot top countries/universities by placement rate
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
sns.barplot(data=origin_placement.head(10), x='placement_rate', y='origin_country', palette='viridis')
plt.title('Top 10 Origin Countries by Placement Rate')
plt.xlabel('Placement Rate')

plt.subplot(1, 3, 2)
sns.barplot(data=destination_placement.head(10), x='placement_rate', y='destination_country', palette='magma')
plt.title('Top 10 Destination Countries by Placement Rate')
plt.xlabel('Placement Rate')

plt.subplot(1, 3, 3)
sns.barplot(data=uni_placement.head(10), x='placement_rate', y='university_name', palette='coolwarm')
plt.title('Top 10 Universities by Placement Rate')
plt.xlabel('Placement Rate')

plt.tight_layout()
plt.show()
```

```python
# Encode target variable (placement_status)
# Encode target: Placed = 1, Not Placed = 0
df['placement_status_encoded'] = df['placement_status'].map({'Placed': 1, 'Not Placed': 0})
```

```python
# Select features
features = [
    'origin_country',
    'destination_country',
    'destination_city',
    'university_name',
    'course_name',
    'field_of_study',
    'year_of_enrollment',
    'scholarship_received',
    'enrollment_reason',
    'graduation_year',
    'gpa_or_score',
    'visa_status',
    'post_graduation_visa',
    'language_proficiency_test',
    'test_score'
]

X = df[features]
y = df['placement_status_encoded']
```

```python
# Encode categorical variables - OneHotEncoding
# One-hot encode categorical columns automatically
X_encoded = pd.get_dummies(X, drop_first=True)
```

```python
# : Split into training and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
# Scale numeric features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Scale only numerical features (we know these two are numeric)
num_cols = ['year_of_enrollment', 'graduation_year', 'gpa_or_score', 'test_score']
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
```

```python
# Logistic Regression model to predict placement_status
# Import & train the model
from sklearn.linear_model import LogisticRegression

# Create and train the model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
# Make predictions
# Predict on test set
y_pred = model.predict(X_test)
```

```python
# Evaluate performance
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Confusion matrix
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Classification report
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# View feature importance
# Logistic regression has coef_ to show which features push predictions toward Placed or Not Placed.
import numpy as np

coefficients = pd.Series(model.coef_[0], index=X_train.columns)
top_positive = coefficients.sort_values(ascending=False).head(10)
top_negative = coefficients.sort_values().head(10)

print("\nTop features pushing toward Placed:\n", top_positive)
print("\nTop features pushing toward Not Placed:\n", top_negative)
```

```python
# Plot ROC Curve
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Predict probabilities
y_probs = model.predict_proba(X_test)[:, 1]

# Compute ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
roc_auc = auc(fpr, tpr)

# Plot
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, label=f'Logistic Regression (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Tune Hyperparameters
from sklearn.model_selection import GridSearchCV

# Create parameter grid
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty': ['l2'],
    'solver': ['liblinear']
}

grid_model = GridSearchCV(LogisticRegression(max_iter=1000), param_grid, cv=5, scoring='accuracy')
grid_model.fit(X_train, y_train)

print("Best Parameters:", grid_model.best_params_)
print("Best Cross-Validated Accuracy:", grid_model.best_score_)
```
