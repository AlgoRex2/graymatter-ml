```python
# import Libraries
from scipy import stats

import numpy as np

import matplotlib.pyplot as plt   # âœ… Import for plotting
```

```python
# Simple Sample Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]
```

```python
# calculation n printing Values
slope, intercept, r, p, std_err = stats.linregress(x, y)
print("Slope:", slope)
print("Intercept:", intercept)
print("R-value:", r)
print("P-value:", p)
print("Standard error:", std_err)

```

```python
# Visualizing the Data
# Plot Graph
# Create model (predicted y values)
mymodel = slope * np.array(x) + intercept

# Plotting
plt.scatter(x, y)                     # Original data
plt.plot(x, mymodel, color='red')    # Regression line
plt.ylim(ymin=0, ymax=2000)
plt.xlim(xmin=0, xmax=200)
plt.xlabel("Average_Pulse")
plt.ylabel("Calorie_Burnage")
plt.title("Linear Regression Example")
plt.grid(True)
plt.show()
```

```python
# Use statsmodels for Detailed Output
import statsmodels.api as sm

# Add constant for intercept
x_with_const = sm.add_constant(x)

# Fit linear regression model
model = sm.OLS(y, x_with_const)
results = model.fit()

# Print full statistical summary
print(results.summary())
```

```python
# Check Confidence Intervals
print(results.conf_int())
```

```python
# Plot and Analyze Residuals
import matplotlib.pyplot as plt

residuals = results.resid
fitted = results.fittedvalues

plt.scatter(fitted, residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.grid(True)
plt.show()
```
