```python
# import Libraries
from scipy import stats

import numpy as np

import matplotlib.pyplot as plt   # âœ… Import for plotting
```

```python
# Simple Sample Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]
```

```python
# calculation n printing Values
slope, intercept, r, p, std_err = stats.linregress(x, y)
print("Slope:", slope)
print("Intercept:", intercept)
print("R-value:", r)
print("P-value:", p)
print("Standard error:", std_err)

```

```python
# Visualizing the Data
# Plot Graph
# Create model (predicted y values)
mymodel = slope * np.array(x) + intercept

# Plotting
plt.scatter(x, y)                     # Original data
plt.plot(x, mymodel, color='red')    # Regression line
plt.ylim(ymin=0, ymax=2000)
plt.xlim(xmin=0, xmax=200)
plt.xlabel("Average_Pulse")
plt.ylabel("Calorie_Burnage")
plt.title("Linear Regression Example")
plt.grid(True)
plt.show()
```
